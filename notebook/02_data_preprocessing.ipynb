## Data Preprocessing :

---------->Data Fetching <-----------

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('D:/ML_Projects\Loan_Approval\data\loan_data.csv')
df
print(df["loan_intent"].unique())
print(df["person_home_ownership"].unique())
print(df["person_education"].unique())
```

----->Label encoding<----

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['person_gender'] = encoder.fit_transform(df['person_gender'])
df['previous_loan_defaults_on_file'] = encoder.fit_transform(df['previous_loan_defaults_on_file'])
categorical_cols = [
    "person_education",
    "person_home_ownership",
    "loan_intent"
]

df = pd.get_dummies(
    df,
    columns=categorical_cols,
    dtype=int,
    drop_first=True
)
df
```

-------> Outliers Handling <-------

------>Age<------

```python
df = df[df['person_age'] < 100]
```

------>person_emp_exp<------

```python
df = df[df['person_emp_exp'] <= 80]
```

```python
df.to_csv("cleaned_loan_data.csv", index=False)
```

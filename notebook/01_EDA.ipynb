--> categorical value <--

```python
# chi2 testing
from scipy.stats import chi2_contingency
alpha = 0.05

for cols in categorical_cols :
    table = pd.crosstab(df[cols],df['loan_status'])
    chi2,p,dof,excepted = chi2_contingency(table)
    if p < alpha:
        print(cols,"Reject H0 (keep feature)")
    else:
        print(cols,"Fail to Reject H0 (remove feature)")
    
```
